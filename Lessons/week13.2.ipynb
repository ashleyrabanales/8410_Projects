```python
import plotly
import plotly.express as px
import pandas as pd
df = pd.read_json("/Users/ashleyrabanales/8410_Projects/Lessons/locations.json")
```

```python
df.head()

fig = px.scatter_mapbox(df,
                        lat = "lat",
                        lon = "lng",
                        hover_name = "name",
                        zoom = 7)
fig.update_layout(mapbox_style = "open-street-map")
fig.update_layout(margin={"r":10, "t":10, "l":10, "b":10})
fig.show()
```
