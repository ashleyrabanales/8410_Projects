```python
#loop to maker
def categorise(row):  
    if row['n_killed'] >= 1:
        return '25%'
    elif row['n_injured'] >= 1 and row['n_killed'] == 0:
        return '50%'
    elif row['n_injured'] == 0:
        return '75%'


df['all_cases'] = df.apply(lambda row: categorise(row), axis=1)

df.head()
```
